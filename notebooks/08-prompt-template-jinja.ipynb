```python
from langchain_core.prompts import PromptTemplate

template = """\
# Instructions for the AI Assistant

Remember that today is {today_date}

## Role
You are a digital marketing expert with 7 years of experience.

## Task
Analyze the following ad text and provide 3 suggestions for improvement, focusing on the Call to Action (CTA).

## Ad Text
{ad_text}

## Output Format
- *Suggestion 1:* [Your first suggestion]
- *Suggestion 2:* [Your second suggestion]
- *Suggestion 3:* [Your third suggestion]
"""

prompt_template = PromptTemplate.from_template(template)
prompt = prompt_template.format(today_date="2025-09-20", ad_text="Este es un anuncio de prueba")
print(prompt)
```

```python
import datetime
today = datetime.datetime.now().strftime("%Y-%m-%d")
prompt_template = PromptTemplate.from_template(template, partial_variables={"today_date": today})
prompt = prompt_template.format(ad_text="This is a test prompt")
print(prompt)
```

```python
template = """\
Your are a helpful that responds to the user's 

{% if name %}
The user's name is {{ name }} and you can call them by that name
{% endif -%}

The user's message is: {{ user_message }}
"""
prompt_template = PromptTemplate.from_template(template, template_format="jinja2")
prompt = prompt_template.format(user_message="Como puedo mejorar mi sitio web?")
print(prompt)
```
