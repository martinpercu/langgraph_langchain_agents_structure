```python
import os

from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-4o-mini", temperature=0)
response = llm.invoke("Hey I'm Martin My contact number is +13478766500")
response.text
```

```python

llm_2 = init_chat_model("anthropic:claude-haiku-4-5-20251001", temperature=0)
response_2 = llm_2.invoke("Hey I'm Martin My contact number is +13478766500")
# response_2.text
response_2.pretty_print()
```

```python
from pydantic import BaseModel, Field

class UserInfo(BaseModel):
    """Contact information for a user."""
    name: str = Field(description="The name of the user")
    email: str = Field(description="The email address of the user")
    phone: str = Field(description="The phone number of the user")
    tone: int = Field(description="The tone of the user", ge=0, le=100)
    age: str = Field(description="The age of the user")
    sentiment: str = Field(description="The sentiment conversation of the user")

llm_with_structured_output  = llm.with_structured_output(schema=UserInfo)

message = [
    ("system", "You are a helpful assistant that can extract contact information from a given conversation."),
    ("user", "Hey I'm Martin My contact number is +13478766500, I'm quite old. I'm almost 50 years old"),
]

response = llm_with_structured_output.invoke(message)
response
```

```python

llm_with_structured_output_2  = llm_2.with_structured_output(schema=UserInfo)

message = [
    ("system", "You are a helpful assistant that can extract contact information from a given conversation."),
    ("user", "Hey I'm Martin My contact number is +13478766500, I'm quite old. I'm almost 50 years old"),
]

response = llm_with_structured_output_2.invoke(message)
response
```
